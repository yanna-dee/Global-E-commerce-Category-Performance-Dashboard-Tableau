```python
#save on a dataframe
df = pd.read_csv('global_ecommerce_sales.csv')
df
```

```python
#check column and data type
df.info()
```

```python
#change column names into snake case format
df.columns = df.columns.str.lower().str.replace(' ', '_')
df
```

```python
#change order_date format and remove all november transactions
df['order_date'] = pd.to_datetime(df['order_date'])
df = df[df['order_date'].dt.month != 11]
display(df)
```

```python
#save into excel file
df.to_excel('global_ecommerce_sales_cleaned.xlsx', index=False)

```
